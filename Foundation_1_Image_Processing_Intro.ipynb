# 🖼️ Image Data Types (dtypes) & Ranges

These ranges define the minimum and maximum values a pixel can hold based on its **bit-depth**.

### **Unsigned Integers (Positive only)**
* **`uint8`**: 0 to 255 (Standard for JPEG/PNG)
* **`uint16`**: 0 to 65,535 (Scientific/Medical imaging)
* **`uint32`**: 0 to $2^{32} - 1$ (High-precision data)

### **Floating Point (Decimals)**
* **`float`**: Usually **0.0 to 1.0** (normalized) or **-1.0 to 1.0**.
    > *Note: Essential for math operations to prevent rounding errors.*

### **Signed Integers (Positive & Negative)**
| Type | Range |
| :--- | :--- |
| **`int8`** | -128 to 127 |
| **`int16`** | -32,768 to 32,767 |
| **`int32`** | $-2^{31}$ to $2^{31} - 1$ |

---

## 🛠️ Scikit-Image Conversion Functions

These functions convert images to a desired `dtype` while **properly rescaling** their values to fit the new range.

* **`img_as_float`**: Converts to **64-bit floating point**.
    * *Effect:* Scales `0-255` (uint8) or `0-65535` (uint16) down to the **0.0 to 1.0** range.
* **`img_as_ubyte`**: Converts to **8-bit unsigned integer**.
    * *Effect:* Scales values (like floats) back to the **0 to 255** range.
* **`img_as_uint`**: Converts to **16-bit unsigned integer**.
    * *Effect:* Scales values to the **0 to 65,535** range.
* **`img_as_int`**: Converts to **16-bit signed integer**.
    * *Effect:* Scales values to the **-32,768 to 32,767** range.

---
> **Pro-Tip:** Always use these functions instead of `.astype()` from NumPy. If you use `.astype(np.uint8)` on a float image, a pixel with value `0.5` becomes `0`. If you use `img_as_ubyte`, it correctly becomes `127`.
